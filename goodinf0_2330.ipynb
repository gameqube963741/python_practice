```python
import requests
headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36 Edg/84.0.522.52'
}
res = requests.get('https://goodinfo.tw/StockInfo/StockBzPerformance.asp?STOCK_ID=2330', headers = headers)
res.encoding = 'utf-8-sig'
# res.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text, 'lxml')
data = soup.select_one('#txtFinDetailData') #是ID 所以給 '#'
```

```python
import pandas
dfs = pandas.read_html(data.prettify())
```
